```python
import os
import cv2
import matplotlib.pyplot as plt

path = os.path.join('resources', 'now_what.jpg')

img = cv2.cvtColor(cv2.imread(path), cv2.COLOR_BGR2HSV)

plt.imshow(cv2.cvtColor(img, cv2.COLOR_HSV2RGB))
plt.show()
```

```python
a = img.shape[0] # height
b = img.shape[1] # width
h, s, v = cv2.split(img)

h[a//4:a//2+a//4, b//4:b//2+b//4] = (h[a//4:a//2+a//4, b//4:b//2+b//4] + 90) % 180

newImg = cv2.merge((h, s, v))

plt.imshow(cv2.cvtColor(newImg, cv2.COLOR_HSV2RGB))
```

```python
from numpy import sqrt

a = img.shape[0] # height
b = img.shape[1] # width
h, s, v = cv2.split(img)

for i in range(a):
    for j in range(b):
        if sqrt((i - a//2)**2 + (j - b//2)**2) < 150:
            h[i, j] = (h[i, j] + 90) % 180

newImg = cv2.merge((h, s, v))

plt.imshow(cv2.cvtColor(newImg, cv2.COLOR_HSV2RGB))
```

```python
import numpy as np

a = img.shape[0] # height
b = img.shape[1] # width
h, s, v = cv2.split(img)

h = h.astype(np.int32) # convert to int32 to avoid overflow

for i in range(a):
    for j in range(b):
        if sqrt((i - 260)**2 + (j - 500)**2) > 200:
            h[i, j] = (h[i, j] + 90) % 180

h = h.astype(np.uint8)

newImg = cv2.merge((h, s, v))

plt.imshow(cv2.cvtColor(newImg, cv2.COLOR_HSV2RGB))
```
