```python
import os
import cv2
import matplotlib.pyplot as plt

path = os.path.join('resources', 'doge.jpg')

img = cv2.imread(path)

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
imgWithLine = img.copy()

pt1 = (1200, 100)
pt2 = (1000, 500)
color = (255, 0, 0)

cv2.line(imgWithLine, pt1, pt2, color[::-1], 5, lineType=cv2.LINE_AA)

plt.imshow(imgWithLine[:, :, ::-1])
```

```python
imgWithCircle = img.copy()

center = (550, 450)
radius = 80
color = (255, 0, 0)

cv2.circle(imgWithCircle, center, radius, color[::-1], -1, lineType=cv2.LINE_AA)

plt.imshow(imgWithCircle[:, :, ::-1])
```

```python
imgWithRectangle = img.copy()

a = (380, 20)
b = (1100, 750)
color = (0, 255, 0)

cv2.rectangle(imgWithRectangle, a, b, color[::-1], 10, lineType=cv2.LINE_AA)

plt.imshow(imgWithRectangle[:, :, ::-1])
```

```python
imgWithText = img.copy()

text = 'much wow'
org = (400, 775)

cv2.putText(imgWithText, text, org, cv2.FONT_HERSHEY_DUPLEX, 4, (255, 0, 0), 7, lineType=cv2.LINE_AA)

plt.imshow(imgWithText[:, :, ::-1])
```

```python
imgCropped = img[:800, 370:1170]

imgResizedByFactors = cv2.resize(imgCropped, None, fx=2, fy=2)

imgResizedBySize = cv2.resize(imgCropped, (10, 10))

plt.figure(figsize=(20, 10))
plt.subplot(131); plt.imshow(imgCropped[:, :, ::-1])
plt.subplot(132); plt.imshow(imgResizedByFactors[:, :, ::-1])
plt.subplot(133); plt.imshow(imgResizedBySize[:, :, ::-1])
```

```python
imgFliped = img.copy()

imgFliped1 = cv2.flip(imgFliped, 1)
imgFliped2 = cv2.flip(imgFliped, 0)
imgFliped3 = cv2.flip(imgFliped, -1)

plt.figure(figsize=(20, 10))
plt.subplot(221); plt.imshow(imgFliped[:, :, ::-1])
plt.subplot(222); plt.imshow(imgFliped1[:, :, ::-1])
plt.subplot(223); plt.imshow(imgFliped2[:, :, ::-1])
plt.subplot(224); plt.imshow(imgFliped3[:, :, ::-1])
```
