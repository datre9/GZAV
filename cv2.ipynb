```python
bearImg = cv2.imread(bearFilePath, 1)

plt.imshow(bearImg)
```

```python
bearImgRGB = cv2.cvtColor(bearImg, cv2.COLOR_BGR2RGB)

plt.imshow(bearImgRGB)
```

```python
r, g, b = cv2.split(bearImgRGB)

plt.figure(figsize=(20, 5))

plt.subplot(141); plt.imshow(r, cmap='gray'); plt.title('Red')
plt.subplot(142); plt.imshow(g, cmap='gray'); plt.title('Green')
plt.subplot(143); plt.imshow(b, cmap='gray'); plt.title('Blue')


imgMerged = cv2.merge((r, g, b))

plt.subplot(144); plt.imshow(imgMerged); plt.title('Original')
```

```python
bearHSV = cv2.cvtColor(bearImgRGB, cv2.COLOR_RGB2HSV)

h, s, v = cv2.split(bearHSV)

plt.figure(figsize=(20, 5))

plt.subplot(141); plt.imshow(h, cmap='gray'); plt.title('Hue')
plt.subplot(142); plt.imshow(s, cmap='gray'); plt.title('Saturation')
plt.subplot(143); plt.imshow(v, cmap='gray'); plt.title('Value')

plt.subplot(144); plt.imshow(cv2.cvtColor(bearHSV, cv2.COLOR_HSV2RGB)); plt.title('Original')
```

```python
h_new = (h + 135) % 180 # opencv uses 0-180 for hue

newMerged = cv2.merge((h_new, s, v))

plt.imshow(cv2.cvtColor(newMerged, cv2.COLOR_HSV2RGB))
```

```python
outPath = os.path.join("output", "bear_hue_shifted.jpg")

cv2.imwrite(outPath, cv2.cvtColor(newMerged, cv2.COLOR_HSV2BGR))
```
